Transforming all yes/no columns to binary values through get_dummies() function.

```python
df[['sex_F','sex_M']] = pd.get_dummies(df[['sex']])
df.head()
```

```python
df[['famsize_LE3','famsize_GT3']] = pd.get_dummies(df[['famsize']])
```

```python
df[['Pstatus_A','Pstatus_T']] = pd.get_dummies(df[['Pstatus']])
```

```python
df[['address_R','address_U']] = pd.get_dummies(df[['address']])
```

```python
df[['schoolsup_NO','schoolsup_YES']] = pd.get_dummies(df[['schoolsup']])
df[['famsup_NO','famsup_YES']] = pd.get_dummies(df[['famsup']])
df[['activities_NO','activities_YES']] = pd.get_dummies(df[['activities']])
df[['paid_NO','paid_YES']] = pd.get_dummies(df[['paid']])
df[['internet_NO','internet_YES']] = pd.get_dummies(df[['internet']])
df[['nursery_NO','nursery_YES']] = pd.get_dummies(df[['nursery']])
df[['higher_NO','higher_YES']] = pd.get_dummies(df[['higher']])
df[['romantic_NO','romantic_YES']] = pd.get_dummies(df[['romantic']])
df[['school_GP','school_MS']] = pd.get_dummies(df[['school']])
```

```python
df.head()
```

We then drop the not needed columns, since the categorical column will give two columns after the binarization of zeros and ones and so one of them is enoughh to deal with.

```python
df1 = df.drop(df.iloc[:, 15:23], inplace = False, axis=1)
```

```python
df1.drop(df1[['romantic_NO']], inplace = True, axis=1)
```

```python
df1.drop(df1[['internet_NO','paid_NO']], inplace = True, axis=1);
```

```python
df1.drop(df1[['higher_NO','nursery_NO','activities_NO','famsup_NO','schoolsup_NO','famsize_LE3','Pstatus_A','address_R','school_MS']], inplace = True, axis=1);
```

```python
df1.drop(df1[['famsize','Pstatus','address','sex','school']], inplace = True, axis=1);
```

```python
df1.head()
print(df1.shape)
```

Encoding categorical features such as father job attribute(Fjob), where {1:at_home, 2: health,
3: services, 4: teacher, 5: other}.

```python
toIntegers = {"Mjob":     {"at_home": 1, "health": 2, "services": 3, "teacher": 4, "other": 5},
              "Fjob": {"at_home": 1, "health": 2, "services": 3, "teacher": 4, "other": 5},
              "reason": {"course": 1, "home": 2, "reputation": 3, "other": 4},
              "guardian": {"father": 1, "mother": 2, "other": 3}}
```

```python
df1.replace(toIntegers, inplace=True)
df1.head()
```

```python
print(df1.shape)
```
